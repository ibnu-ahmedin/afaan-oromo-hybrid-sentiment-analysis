```python
# ==========================================================
# 🔥 1. IMPORTS
# ==========================================================
import pandas as pd
from sklearn.model_selection import train_test_split

# ==========================================================
# 🔥 2. LOAD RAW DATA (ONLY ONCE)
# ==========================================================
df = pd.read_csv("/content/Full_Corpus_10005.csv")

# Clean labels
df = df.dropna(subset=["Text", "Sentiment"]).reset_index(drop=True)
df["Sentiment"] = df["Sentiment"].str.strip().str.capitalize()

# Label encoding
label_map = {"Negative": 0, "Neutral": 1, "Positive": 2}
df["label"] = df["Sentiment"].map(label_map)

print("Dataset size:", df.shape)

# ==========================================================
# 🔥 3. STRATIFIED SPLIT (FIXED & REPRODUCIBLE)
# ==========================================================
# 80% train, 10% val, 10% test

train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label"],
    random_state=42
)

print("\nSplit sizes:")
print("Train:", train_df.shape)
print("Validation:", val_df.shape)
print("Test:", test_df.shape)

# ==========================================================
# 🔥 4. SAVE SPLITS (CRITICAL FOR CONSISTENCY)
# ==========================================================
train_df.to_csv("/content/train.csv", index=False)
val_df.to_csv("/content/val.csv", index=False)
test_df.to_csv("/content/test.csv", index=False)

print("\n✅ Saved:")
print(" - train.csv")
print(" - val.csv")
print(" - test.csv")
```
