```python
# =====================================
# MCNEMAR TEST
# =====================================

!pip install statsmodels -q

import pandas as pd
import numpy as np
from statsmodels.stats.contingency_tables import mcnemar

# Load predictions
df_h = pd.read_csv("/content/hybrid_preds.csv")
df_a = pd.read_csv("/content/afri_preds.csv")

# Extract values
y_true = df_h["y_true"].values
hybrid_preds = df_h["hybrid_preds"].values
afri_preds = df_a["afri_preds"].values

# Sanity check (VERY IMPORTANT)
print("Length check:", len(y_true), len(hybrid_preds), len(afri_preds))

# Build contingency table
n01 = np.sum((hybrid_preds == y_true) & (afri_preds != y_true))
n10 = np.sum((hybrid_preds != y_true) & (afri_preds == y_true))

table = [[0, n01],
         [n10, 0]]

# Run McNemar
result = mcnemar(table, exact=False, correction=True)

print("\n======================================")
print("McNemar Test: Hybrid vs AfriBERTa")
print("======================================")
print("n01 (Hybrid correct, Afri wrong):", n01)
print("n10 (Hybrid wrong, Afri correct):", n10)
print("Statistic:", round(result.statistic, 4))
print("p-value:", result.pvalue)

# Interpretation
if result.pvalue < 0.05:
    print("\n✅ Statistically significant difference (Hybrid is better)")
else:
    print("\n❌ No statistically significant difference")
```
